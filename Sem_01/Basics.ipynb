### Going further

```
# Messing everything around.
img = cv2.imread('/content/drive/MyDrive/MachineLearning/machine_learning/Sem_01/lena.jpg' )
# Load the image in color

# Convert the color image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using matplotlib
plt.imshow(img_gray, cmap='terrain')
plt.axis('off')  # Optional: Turn off the axis
plt.show()
```

```
# Playing with image intensities
img = cv2.imread('/content/drive/MyDrive/MachineLearning/machine_learning/Sem_01/lena.jpg' )

# Iterate through rows and columns of the image
for r in range(img_gray.shape[0]):  # Go through rows
    for c in range(img_gray.shape[1]):  # Go through columns
        if img_gray[r, c] < 128:  # Example condition: if the pixel intensity is less than 128
            img_gray[r, c] = 0  # Set the pixel to black
        else:
            img_gray[r, c] = 255  # Set the pixel to white

# Display the modified image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Optional: Turn off the axis
plt.show()
```

### Save an image using OpenCV

### Crop the image - taking regions of interest (ROI)

```
# Extract some ROI

img = cv2.imread('/content/drive/MyDrive/MachineLearning/machine_learning/Sem_01/lena.jpg' )

roi = img[50:150, 100:200]  # Example ROI from (x1=100, y1=50) to (x2=200, y2=150)

# Display the ROI
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB if the image is in color
plt.axis('off')  # Optional: Turn off the axis
plt.show()
```

```
img_gray = cv2.imread('/content/drive/MyDrive/MachineLearning/machine_learning/Sem_01/lena.jpg', cv2.IMREAD_GRAYSCALE )

roi = img_gray[50:150, 100:200]  # Example ROI from (y1=50, x1=100) to (y2=150, x2=200)

# Place the ROI back at a different location in the image
img_gray[200:300, 250:350] = roi  # Replace part of the image with the ROI

# Display the modified image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```
